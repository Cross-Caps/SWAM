**Geometric** **Properties**

```python
def get_metrics_single_res(checkpoint_path):
    model_state_dict = torch.load(checkpoint_path, map_location = "cpu")
    layer_1_weights = model_state_dict["seq_net.conv_1.weight"]
    layer_2_weights = model_state_dict["seq_net.conv_2.weight"]
    layer_3_weights = model_state_dict["seq_net.conv_3.weight"]
    layer_1_weights = layer_1_weights.view(80, -1).numpy().T
    layer_2_weights = layer_2_weights.view(100, -1).numpy().T
    layer_3_weights = layer_3_weights.view(100, -1).numpy().T
    return layer_1_weights, layer_2_weights, layer_3_weights
```

Multi-Resolution Model

```python
def get_simple_padded_matrix(lay_names, model_state_dict, k):
    int_list = []
    ma = -1
    for nam in lay_names:
        u = model_state_dict[nam]
        ma = max(u.size()[2], ma)
    for nam in lay_names:
        u = model_state_dict[nam]
        a, b, c = u.size()
        int_list.append(torch.cat([u, torch.zeros((a, b, ma - c))], axis = 2))
    for i, u in enumerate(int_list):
        int_list[i] = u.view(k, -1)
    return torch.cat(int_list, axis = 0)

def get_metrics_multi_res(checkpoint_path):
    model_state_dict = torch.load(checkpoint_path, map_location = "cpu")
    lay_1_names = ['layer_1_1.weight', 'layer_1_2.weight', 'layer_1_3.weight', 'layer_1_4.weight', 'layer_1_5.weight', 'layer_1_6.weight', 'layer_1_7.weight', 'layer_1_8.weight']
    lay_2_names = ['layer_2_1.weight', 'layer_2_2.weight', 'layer_2_3.weight', 'layer_2_4.weight', 'layer_2_5.weight']
    lay_3_names = ['layer_3_1.weight', 'layer_3_2.weight', 'layer_3_3.weight', 'layer_3_4.weight']
    layer_1_weights = get_simple_padded_matrix(lay_1_names, model_state_dict, 10).numpy().T
    layer_2_weights = get_simple_padded_matrix(lay_2_names, model_state_dict, 20).numpy().T
    layer_3_weights = get_simple_padded_matrix(lay_3_names, model_state_dict, 25).numpy().T
    return layer_1_weights, layer_2_weights, layer_3_weights
```

```python
model = "google_assistant_dataset_multifeat_final_model.pt"
layer_1, layer_2, layer_3 = get_metrics_multi_res(model)
```

```python
D = DictionaryMetrics(layer_1, gap = None, rad = None)
print("Layer 1 Metrics")
print(D)
```

```python
D = DictionaryMetrics(layer_2, gap = None, rad = None)
print("Layer 2 Metrics")
print(D)
```
